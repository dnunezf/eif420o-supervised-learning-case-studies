```python
# Formaremos el d', normalizando el dataset

import pandas as pd
import numpy as np
import seaborn as sns                       #visualisation
import matplotlib.pyplot as plt             #visualisation
%matplotlib inline     
sns.set(color_codes=True)

df = pd.read_csv("/home/dnunezf/Documents/AI_LABS/Lab_26-02-2026/data.csv")
# To display the top 5 rows 
df.head(5)          
```

```python
df.tail(5)
```

```python
df.dtypes # consultar los tipos de datos
```

```python
# eliminar columnas irrelevantes
df = df.drop(['Engine Fuel Type', 'Market Category', 'Vehicle Style', 'Popularity', 'Number of Doors', 'Vehicle Size'], axis=1)
df.head(5)
```

```python
# renombrar columnas
df = df.rename(columns={"Engine HP": "HP", "Engine Cylinders": "Cylinders", "Transmission Type": "Transmission", "Driven_Wheels": "Drive Mode","highway MPG": "MPG-H", "city mpg": "MPG-C", "MSRP": "Price" })
df.head(5)
```

```python
df.shape # individuos, variables
```

```python
# extaer individuos duplicados (filas duplicadas)
duplicate_rows_df = df[df.duplicated()]
print("number of duplicate rows: ", duplicate_rows_df.shape)
```

```python
df.count()      # Used to count the number of rows
```

```python
df = df.drop_duplicates() # eliminar duplicado y dejarlos en el df
df.head(5)
```

```python
df.count() # tras eliminar los duplicados
```

```python
# eliminar valores nulos o faltantes

print(df.isnull().sum()) # sumariza los valores nulos por columnas, a nivel de todos los individuos
```

```python
df = df.dropna()    # Dropping the missing values.
df.count()
```

```python
print(df.isnull().sum())   # After dropping the values
```

```python
# Deteccion de atipicos en alguna variable de interes

sns.boxplot(x=df['Price'])
```

```python
sns.boxplot(x=df['HP'])
```

```python
sns.boxplot(x=df['Cylinders'])
```

```python
# se maneja el criterio de los cuartiles
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
df = df[~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
df.shape
```

```python
# Ploteos

# Histograma

df.Make.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Number of cars by make")
plt.ylabel('Number of cars')
plt.xlabel('Make');
```

```python
# Heat Maps

plt.figure(figsize=(10,5))
c= df.corr() # retorna la matriz de correlacion
sns.heatmap(c,cmap="BrBG",annot=True)
c
```

```python
# Scatterplot

fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(df['HP'], df['Price'])
ax.set_xlabel('HP')
ax.set_ylabel('Price')
plt.show()
```
