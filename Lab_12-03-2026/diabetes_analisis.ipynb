```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score
from sklearn.preprocessing import StandardScaler

# ===============================
# 1. Cargar dataset
# ===============================

X = df.drop("Defective", axis=1)
y = df["Defective"]

# ===============================
# 2. Escalar variables
# ===============================

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ===============================
# 3. Train / Test split
# ===============================

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y,
    test_size=0.3,
    random_state=42,
    stratify=y
)

# ===============================
# 4. Logistic Regression (None)
# ===============================

lr_none = LogisticRegression(
    max_iter=1000,
    class_weight=None
)

lr_none.fit(X_train, y_train)

proba_none = lr_none.predict_proba(X_test)[:,1]

auc_none = roc_auc_score(y_test, proba_none)

# ===============================
# 5. Logistic Regression (Balanced)
# ===============================

lr_balanced = LogisticRegression(
    max_iter=1000,
    class_weight='balanced'
)

lr_balanced.fit(X_train, y_train)

proba_balanced = lr_balanced.predict_proba(X_test)[:,1]

auc_balanced = roc_auc_score(y_test, proba_balanced)

# ===============================
# 6. Benchmark
# ===============================

results = pd.DataFrame({
    "Model": ["Logistic Regression (None)", "Logistic Regression (Balanced)"],
    "AUC": [auc_none, auc_balanced]
})

print(results)
```
